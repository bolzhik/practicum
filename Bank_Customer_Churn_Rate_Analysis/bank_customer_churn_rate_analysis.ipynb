### Analysis of the features distribution

```python
def find_whiskers(column):
    '''
    the function gets a column name as input and returns left and right whiskers
    '''    
    stat = data[column].describe() 
    # the interquartile range
    iqr = stat[6] - stat[4] #stat[6] - 75%, stat[4] - 25%    
    # left and right whiskers
    left_whisker = round(stat[4] - 1.5 * iqr, 2)
    right_whisker = round(stat[6] + 1.5 * iqr, 2)
    # left whisker should not be less than the minimum value
    if left_whisker < stat[3]:
        left_whisker = stat[3]
        
    # right whisker should not be greater than the maximum value
    if right_whisker > stat[7]:
        right_whisker = stat[7]
        
    return [left_whisker, right_whisker]
```

```python
def find_anomalies_ratio(column):
    '''
    the function receives a column name as input 
    and returns the proportion of abnormal values from the entire dataset
    '''
    return 'Abnormal values are {:.1%}'.format(len(
        data[(data[column] > find_whiskers(column)[1]) | (data[column] < find_whiskers(column)[0])]) 
                                                          / len(data)) + ' from the entire dataset'
```

```python
def box_and_hist_plot(data, column, right, left=0, bins='auto', cut=False):
    '''
    The function plots a boxplot and a normalized histogram for a column.
    
    The function receives a dataset, a column name, limits for the x-axis, 
    bins for the histogram and cut condition for deleting outliers from the dataset.
    '''
    plt.figure(figsize = (20,15))
    
    plt.subplot(2, 1, 1)
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot for {column_quant_dict_title[column]}')
    plt.xlabel(f'{column_quant_dict_label[column]}')
    plt.xlim(left, right)
    plt.subplots_adjust(hspace=0.5)
    
    plt.subplot(2, 1, 2)
    if cut:
        data=data[(data[column] > find_whiskers(column)[0]) & 
                          (data[column] <= find_whiskers(column)[1])]
    sns.histplot(x=column,
                data=data,
                bins=bins,
                hue='churn',
                stat='probability',
                common_norm=False,
                palette=['#2677bb', '#f67944'])
    plt.title(f'Histogram for {column_quant_dict_title[column]}')
    plt.legend(['lost client', 'current client'])
    plt.xlabel(f'{column_quant_dict_label[column]}')
    plt.xlim(left, right)
    plt.ylabel('Frequency')
    plt.show()
```

```python
def plot_hist_comparison(data, column, bins='auto'):
    '''
    The function plots normalized histograms for a column by churn and not churn (current)
    '''
    sns.histplot(x=column,
                data=data,
                bins=bins,
                hue='churn',
                stat='probability',
                common_norm=False,
                palette=['#2677bb', '#f67944'])
    plt.title(f'The distribution of the {column_quant_dict_title[column]}')
    plt.legend(['lost client', 'current client'])
    plt.xlabel(f'{column_quant_dict_label[column]}')    
    plt.ylabel('Frequency')
    plt.show()
```

```python
# create a dictionary for the chart name
column_quant_dict_title = {'score': 'credit score', 'age': 'age', 'gender_category': 'gender', 
                           'credit_card': 'credit card', 'last_activity': 'last activity', 'segments': 'segments',
                           'balance': 'account balance', 'est_salary': "client's salary", 'city_category': 'city', 
                           'score_category': 'credit score category', 'salary_category': 'salary category',
                           'age_category': 'age category', 'balance_category': 'account balance category', 
                           'churn': 'churn', 'products': 'number of products', 'product_category': 'number of products',
                           'equity': 'number of properties', 'equity_category': 'number of properties'}
```

```python
# plotting barchart for category
def bar_by_category_plot(df, columns=['gender_category', 'credit_card', 'last_activity',
                                  'city_category', 'score_category', 'salary_category', 
                                  'age_category', 'balance_category', 'product_category', 'equity_category'], 
                         hue='churn'):
    data=df.copy()
    data['credit_card'] = data['credit_card'].map({1:'yes', 0:'no'})
    data['last_activity'] = data['last_activity'].map({1:'yes', 0:'no'})
    data['churn'] = data['churn'].map({1:'yes', 0:'no'})
    
    for column in columns:
        plt.figure(figsize=(15, 8))                       
        ax = sns.countplot(x = data[column], hue=hue, data=data, 
                           order=sorted(data[column].unique()), hue_order=['no', 'yes'])
        sns.set_palette('Set2')
        bars = ax.patches
        half = int(len(bars)/2)
        # add percentage to the top of each bar
        for left, right in zip(bars[:half], bars[half:]):
            height_l = left.get_height()
            height_r = right.get_height()
            if np.isnan(height_l):
                height_l = 0
            if np.isnan(height_r):
                height_r = 0
            total = height_l + height_r            
            if total==0:
                total=1                        
            ax.text(left.get_x() + left.get_width()/2., height_l + 40, '{0:.1%}'.format(height_l/total), ha="center", size='small')
            ax.text(right.get_x() + right.get_width()/2., height_r + 40, '{0:.1%}'.format(height_r/total), ha="center", size='small')

        plt.title(f'Grouped bar chart for {column_quant_dict_title[column]}')

        if column == 'balance_category':
            ax.tick_params(labelsize=16)        

        plt.xlabel(f'{column_quant_dict_label[column]}')    
        plt.ylabel('Frequency')
        plt.subplots_adjust(hspace=0.5)
        plt.show()
```

## Hypothesis testing

- Hypothesis about the difference in the churn rate between customers from Yaroslavl and Rostov

### Hypothesis about the difference in income between the customers who left and those who stayed

```python
column='est_salary'
sns.histplot(x=data[column], kde=True)
plt.title(f'Histogram for {column_quant_dict_title[column]}')
plt.xlabel(f'{column_quant_dict_label[column]}')
plt.ylabel('Frequency')
plt.show()
```

We can say that the distribution is not perfect, but close to normal and with a very long right tail. Now let's look at the average and median income by client groups (lost and current).

```python
data.groupby('churn').agg({'user_id': 'count', 'est_salary': [np.mean, np.var]}).style.format('{:.0f}')
```

The average incomes of the two client groups are almost the same. The sizes of the groups differ, as well as the sample variances. There are also outliers in the data. Too many facts are not in favor of Student's t-test, so we will test the hypothesis of equality of means using the Wilcoxon-Mann-Whitney method. It is robust against outliers and works with ranks instead of values. The test will be used for independent samples, and it will be two-tailed.

**Null Hypothesis**: The average incomes of customers who left and customers who stayed are the same.

**Alternative Hypothesis**: The average incomes of customers who left and customers who stayed are different.

The critical level of statistical significance is taken as 5%.

```python
alpha = .05 # critical level of statistical significance

salary_churn_1 = data[data.churn==1]['est_salary']
salary_churn_0 = data[data.churn==0]['est_salary']

p_value = stats.mannwhitneyu(salary_churn_1, salary_churn_0, alternative='two-sided')[1]

print("P-value = {0:.5f}".format(p_value))

if p_value < alpha:
    print("Reject the null hypothesis: there is a significant difference.")
else:
    print("Failed to reject the null hypothesis: there is no significant difference.")
```

**Conclusion**

According to the data at a significance level of 5%, there is a reason to reject the null hypothesis in favor of the alternative one. On average, the incomes of lost and current customers differ.

### Hypothesis about the difference in the churn rate between customers from Yaroslavl and Rostov

In research, it was found that there is no significant difference in outflow between cities, and that the city is not an important feature. Let's look at the churn rate by city.

```python
data.city_category.unique()
```

```python
data[(data.city_category=='Yaroslavl') | (data.city_category=='Rostov')] \
    .groupby('city_category') \
    .agg({'churn': ['count', 'mean']}) \
    .style.format({('churn', 'mean'):'{:.1%}'})
```

We will test the hypothesis using the method of testing proportions based on the normal z-test. The test will be one-sided.

**Null hypothesis**: Among the clients who left, the proportions of clients from Yaroslavl and Rostov are the same.

**Alternative hypothesis**: Among the clients who left, the proportion of clients from Yaroslavl is greater than the proportion from Rostov.

The critical level of statistical significance is taken as 5%.

```python
alpha = .05 # critical level of statistical significance

successes = [
    data[(data.city_category=='Yaroslavl') & (data.churn==1)]['user_id'].count(),
    data[(data.city_category=='Rostov') & (data.churn==1)]['user_id'].count()
]

trials = [
    data[data.city_category=='Yaroslavl']['user_id'].count(),
    data[data.city_category=='Rostov']['user_id'].count()
]

stat, p_value = proportions_ztest(successes, trials, alpha, alternative='larger')

print("P-value = {0:.5f}".format(p_value))

if p_value < alpha:
    print('Reject the null hypothesis: there is a significant difference between the proportions.')
else:
    print('Failed to reject the null hypothesis, there is no significant difference between the proportions.') 
```

**Conclusion**

According to the data at the 5% significance level, there is no reason to reject the null hypothesis in favor of the alternative one. This means that clients from Yaroslavl and Rostov have the same churn rate.
